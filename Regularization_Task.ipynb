# Import Libraries

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
```

# Import Dataset

```python
data = fetch_california_housing()
data
```

```python
X = pd.DataFrame(data = data.data,columns = data.feature_names)
y = data.target
```

```python
X
```

```python
X = X.head(350)
y = y[:350]
```

#  Data Understanding

```python
X.shape,y.shape
```

# Model Building

```python
linear_model = LinearRegression()
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=0.1)
elastic_model = ElasticNet(alpha=0.1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=123,shuffle=True)
```

# Model Training

```python
linear_model.fit(X_train,y_train)
ridge_model.fit(X_train,y_train)
lasso_model.fit(X_train,y_train)
elastic_model.fit(X_train,y_train)
```

# Model Co-efficient

```python
print("Linear Regression Coefficient")
print(linear_model.coef_)

print("Ridge Regression Coefficient")
print(ridge_model.coef_)

print("Lasso Regression Coefficient")
print(lasso_model.coef_)

print("ElasticNet Regression Coefficient")
print(elastic_model.coef_)
```

# Visualization

```python
from matplotlib import pyplot as plt
```

```python
def plot_coefficients(model,feature_names,title):

    coeff_df = pd.DataFrame({"Features": feature_names, "Coefficient":model.coef_})
    
    coeff_df = coeff_df.sort_values(by = "Coefficient")
    print(coeff_df)

    plt.figure(figsize = (10,6))
    plt.barh(y = coeff_df["Features"],width =coeff_df["Coefficient"])

    plt.title(title)
    plt.xlabel("Coefficient_Value")
    plt.ylabel("Features")

    plt.tight_layout()    #Adjust the padding between and around subplots
    plt.show()

#Linear Model
plot_coefficients(linear_model,X.columns,"Linear Regression Coefficient")

#Ridge Model
plot_coefficients(ridge_model,X.columns,"Ridge Regression Coefficient")

#Lasso Model
plot_coefficients(lasso_model,X.columns,"Lasso Regression Coefficient")

#ElasticNet Model
plot_coefficients(elastic_model,X.columns,"ElasticNet Regression Coefficient")
```

# Error Comparison

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
model = {"Linear":linear_model,
         "Ridge":ridge_model,
         "Lasso":lasso_model,
         "ElasticNet":elastic_model}

for name,model in model.items():

    y_pred_test = model.predict(X_test)
    
    mse = mean_squared_error(y_true = y_test ,y_pred = y_pred_test)
    r2 = r2_score(y_test,y_pred_test)
    
    print(name)
    print(mse)
    print(r2)
    print()
```
