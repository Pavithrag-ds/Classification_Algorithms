## 1. Import Necessary Libraries

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

## 2. Import Dataset

```python
wine_data = load_wine()
wine_data_df = pd.DataFrame(wine_data.data,columns = wine_data.feature_names)
wine_data_df["target"] = wine_data.target
wine_data_df
```

## 4. Model Building

```python
X = wine_data_df.drop("target",axis = 1)
y = wine_data_df["target"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=11,shuffle=True,stratify=y)
```

# How to choose the Optimal no.of neighbours?

#### Steps we are going to do:    
     
     - for loop 3-51
     - KNN(n_neighbors=i)
     - Cross Validation
     - Mean
     - Store it in a list

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kfold = KFold(n_splits=5,shuffle=True, random_state=123)
cv_scores_mean = []

for i in range(3,52,2):    # our goal is to get odd numbers..so we set 2 which reflects as(i+2) instead (i+1)-default
     knn_classifier = KNeighborsClassifier(n_neighbors=i)
     knn_classifier.fit(scaled_X,y)
     cv_scores = cross_val_score(estimator = knn_classifier, X = scaled_X, y=y,cv=kfold)
     cv_scores_mean.append(cv_scores.mean().round(4)) #It calculates the mean of i's in each fold(totally 5 folds) & append it in list(cv_scores_mean)
```

```python
cv_scores_mean
```

```python
max(cv_scores_mean)
```

```python
cv_scores_mean.index(max(cv_scores_mean))   #To find the maximum mean given KNN 

#Max value iruka index(row) number of cv_scores_mean
```

#### KNN in 10th Index

### Let's visualize this

```python
from matplotlib import pyplot as plt
plt.plot(range(3,52,2),cv_scores_mean)
plt.xlabel("No.of Neighbours")
plt.ylabel("Mean CV Score")
plt.show()
```

```python
1 = 3    # 1st index value = 3

10* = 23    # 10th index value = 23
```

#### So if we choose K-value = 23...we get the maximum accuracy score...which is the Optimal KNN value

```python
knn_model = KNeighborsClassifier(n_neighbors=23)
knn_model.fit(X_train,y_train)
y_pred_train = knn_model.predict(X_train)
print("Training Accuracy:",round(accuracy_score(y_train,y_pred_train),4))
```
